#### Question 9
Create a pivot table of the average price broken down by manufacturer and RAM size.

```python
laptops_pivot = pd.pivot_table (
    laptops, 
    index = 'RAM_GB', 
    columns = 'Manufacturer', 
    values = 'Price', 
    aggfunc = 'mean'
)
laptops_pivot
```

#### Question 10
Assume that the price in this dataset is in US dollars. Create a new column that represents the price of these laptops in Euros. Assume that 1 USD is equivalent to 0.93 Euros. Add this column to the dataframe. 

```python
def convert (price): 
    price = price * 0.93
    return float(price)
laptops ['Price'] = laptops ['Price'].apply(convert)
```
